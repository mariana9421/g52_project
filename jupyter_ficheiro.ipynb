```python
import pandas as pd
import sqlite3
import os

csv_path = "data/marketplace.csv"
db_path = "marketplace.db"


if os.path.exists(db_path):
    os.remove(db_path)

try:
   
    df = pd.read_csv(csv_path)

    df.columns = df.columns.str.strip()

    category = df[['category_id', 'category_name', 'category_comments']].drop_duplicates()

    conn = sqlite3.connect(db_path)

    category.to_sql('category', conn, if_exists='replace', index=False)
    
    print("Processo concluído com sucesso!")
    print(f"Tabela 'category' criada no arquivo: {db_path}")

except Exception as e:
    print(f"Ocorreu um erro durante a execução: {e}")

finally:
    
    if 'conn' in locals():
        conn.close()
        print("Conexão com o banco de dados fechada.")
```
