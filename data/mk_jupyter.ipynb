```c++
import pandas as pd
import sqlite3


df = pd.read_csv("data/marketplace.csv")


conn = sqlite3.connect("data/marketplace.db")


    
category = df[['category_id', 'category_name', 'category_comments']].drop_duplicates()
category.to_sql("category", conn, if_exists="replace", index=False)


seller = df[['seller_id', 'seller_name', 'seller_address']].drop_duplicates()
seller.to_sql("seller", conn, if_exists="replace", index=False)



marketplace = df[['marketplace_id', 'marketplace_name', 'mkt_created_date',
                  'category_id', 'seller_id']].drop_duplicates()
marketplace.to_sql("marketplace", conn, if_exists="replace", index=False)


transaction = df[['transaction_date', 'marketplace_id', 'seller_id']].drop_duplicates()
transaction.to_sql("transaction", conn, if_exists="replace", index=False)

conn.close()


```
