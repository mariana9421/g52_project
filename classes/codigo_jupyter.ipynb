```c++
import pandas as pd
import sqlite3


df = pd.read_csv("data/marketplace.csv")

conn = sqlite3.connect("data/marketplace.db")

category = (
    df[['category_id', 'category_name', 'category_comments']]
    .drop_duplicates()
    .rename(columns={
        'category_id': 'id',
        'category_name': 'name',
        'category_comments': 'comments'
    })
)

category.to_sql('Category', conn, if_exists='replace', index=False)



seller = (
    df[['seller_id', 'seller_name', 'seller_address']]
    .drop_duplicates()
    .rename(columns={
        'seller_id': 'id',
        'seller_name': 'name',
        'seller_address': 'address'
    })
)

seller.to_sql('Seller', conn, if_exists='replace', index=False)



marketplace = (
    df[['marketplace_id', 'marketplace_name', 'mkt_created_date', 'category_id']]
    .drop_duplicates()
    .rename(columns={
        'marketplace_id': 'id',
        'marketplace_name': 'name',
        'mkt_created_date': 'created_date'
    })
)

marketplace.to_sql('Marketplace', conn, if_exists='replace', index=False)



transaction = (
    df[['transaction_id', 'transaction_date', 'transactions', 'seller_id', 'marketplace_id']]
    .drop_duplicates()
    .rename(columns={
        'transaction_id': 'id',
        'transaction_date': 'date'
    })
)

transaction.to_sql('Transaction', conn, if_exists='replace', index=False)



conn.close()

print("Base de dados criada com sucesso!")


```
